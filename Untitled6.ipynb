```python
%%bash
cat > main.cpp <<EOF
#include <iostream>
#include <thread>
#include <vector>
#include <mutex>
#include <chrono>

int sum = 0;
std::mutex mtx;

// 1) 동기화가 없는 함수 (탐구 6-1: 결과가 200이 안 될 수 있음)
void unsafe_add() {
    for (int i = 0; i < 100000; ++i) {
        sum += 1;
    }
}

// 2) 뮤텍스를 사용한 함수 (탐구 6-2: 결과가 정확함)
void safe_add() {
    for (int i = 0; i < 100000; ++i) {
        mtx.lock();   // 임계 구역 잠금
        sum += 1;
        mtx.unlock(); // 임계 구역 해제
    }
}

int main() {
    // 동기화 미적용 테스트
    sum = 0;
    std::thread t1(unsafe_add);
    std::thread t2(unsafe_add);
    t1.join(); t2.join();
    std::cout << "Unsafe Sum: " << sum << " (Expected 200000)" << std::endl;

    // 뮤텍스 적용 테스트
    sum = 0;
    auto start = std::chrono::high_resolution_clock::now();
    std::thread t3(safe_add);
    std::thread t4(safe_add);
    t3.join(); t4.join();
    auto end = std::chrono::high_resolution_clock::now();

    std::chrono::duration<double> elapsed = end - start;
    std::cout << "Safe Sum: " << sum << std::endl;
    std::cout << "Time taken: " << elapsed.count() << "s (탐구 6-3: 성능 저하 확인)" << std::endl;

    return 0;
}
EOF

g++ -std=c++11 -pthread -o main main.cpp
./main
```

```python
%%bash
cat > main_semaphore.cpp <<EOF
#include <iostream>
#include <thread>
#include <vector>
#include <semaphore> // C++20 기준
#include <queue>
#include <chrono> // std::chrono::milliseconds 를 위해 추가

std::queue<int> buffer;
const int BUFFER_SIZE = 5;

// 탐구 6-4: 세마포어 원리 적용
std::counting_semaphore<5> empty_slots(BUFFER_SIZE); // 비어있는 공간 수
std::counting_semaphore<5> full_slots(0);           // 채워진 데이터 수
std::mutex buffer_mtx;                              // 버퍼 접근 보호 (Mutex)

void producer(int id) {
    for (int i = 0; i < 10; ++i) {
        empty_slots.acquire(); // wait(empty): 빈 공간이 생길 때까지 대기

        buffer_mtx.lock();
        buffer.push(i);
        std::cout << "생산자 " << id << ": " << i << " 생산" << std::endl;
        buffer_mtx.unlock();

        full_slots.release();  // signal(full): 데이터가 들어왔음을 알림
        std::this_thread::sleep_for(std::chrono::milliseconds(100));
    }
}

void consumer(int id) {
    for (int i = 0; i < 10; ++i) {
        full_slots.acquire();  // wait(full): 데이터가 생길 때까지 대기

        buffer_mtx.lock();
        int data = buffer.front();
        buffer.pop();
        std::cout << "소비자 " << id << ": " << data << " 소비" << std::endl;
        buffer_mtx.unlock();

        empty_slots.release(); // signal(empty): 빈 공간이 생겼음을 알림
        std::this_thread::sleep_for(std::chrono::milliseconds(150));
    }
}

int main() {
    std::thread p1(producer, 1);
    std::thread c1(consumer, 1);

    p1.join();
    c1.join();

    return 0;
}
EOF

g++ -std=c++20 -pthread -o main_semaphore main_semaphore.cpp
./main_semaphore
```
